```python
d = r"C:\Users\Nathan\pytorch practice\diabetes.csv" # data file
import torch
from torch.autograd import Variable
import numpy as np
import torch.nn.functional as F

xy = np.loadtxt(d, delimiter = ',', dtype = np.float32) # use numpy to load data from csv as a float32, delimiter as ','
x_data = Variable(torch.from_numpy(xy[:, 0:-1])) # get data from numpy all rows exclude last column(true output)
y_data = Variable(torch.from_numpy(xy[:, [-1]])) # get true output(diabetes or not)


class Model(torch.nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.l1 = torch.nn.Linear(8, 6) # create a linear model with input of 8 features and 1 label, output size of 6
        self.l2 = torch.nn.Linear(6, 4) 
        self.l3 = torch.nn.Linear(4, 1) # down to one output
        
        self.sigmoid = torch.nn.Sigmoid() # gets value between 0 and 1, true or false
        
    def forward(self, x):
        out1 = F.relu(self.l1(x)) # Rectified Linear Unit activation function
        out2 = F.relu(self.l2(out1))
        y_pred = self.sigmoid(self.l3(out2))
        return y_pred
    
model = Model()

criterion = torch.nn.BCELoss(size_average = True) # Binary Cross Entropy Loss function
optimizer = torch.optim.SGD(model.parameters(), lr = 0.1) # Stochastic Gradient Descent, learning rate of 0.1

for epoch in range(1000):
    y_pred = model(x_data)
    
    loss = criterion(y_pred, y_data)
    print(epoch, loss.item())
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```
