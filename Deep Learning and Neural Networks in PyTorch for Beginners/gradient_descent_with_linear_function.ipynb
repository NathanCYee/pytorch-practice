```python
import torch
from torch.autograd import Variable

x_data = [1.0, 2.0, 3.0] # define your x data
y_data = [2.0, 4.0, 6.0] # define your y data

w = Variable(torch.Tensor([1.0]), requires_grad=True) # create your weight

def forward(x): # forward linear function
    return  x * w

def loss(x,y): # calculate loss between predicted - actual (r2)
    y_pred = forward(x)
    return (y_pred-y) * (y_pred-y)

print("My prediction before training", 4, forward(4).data[0]) # get first element of your prediction data

for epoch in range(50): # 10 epochs to adjust weight
    for x_val, y_val, in zip(x_data, y_data):
        l = loss(x_val, y_val)
        l.backward() # gradient descent function (derivative to find slope)
        print("\tgrad: ", x_val, y_val, w.grad.data[0])
        w.data = w.data - 0.01 * w.grad.data # adjust weight by gradient for next epoch
        
        w.grad.data.zero_() # zero your gradient function
    print("my progress:", epoch, l.data[0]) 

print("my prediction after training", 4, forward(4).data[0]) # get first element of your prediction data
```
