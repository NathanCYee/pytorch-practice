```python
import torch
from torch.autograd import Variable

x_data = Variable(torch.Tensor([[1.0], [2.0], [3.0]])) #Create tensors and wrap into variable
y_data = Variable(torch.Tensor([[2.0], [4.0], [6.0]]))

class Model(torch.nn.Module): # create a pytorch Model
    def __init__(self):
        super(Model, self).__init__()
        self.linear = torch.nn.Linear(1, 1) # create a linear regression function with 1, 1 variables
        
    def forward(self, x): # function to run the forward linear function
        y_pred = self.linear(x)
        return y_pred
    
model = Model()

criterion = torch.nn.MSELoss(size_average=False) # loss function(mean squared error)
optimizer = torch.optim.SGD(model.parameters(), lr=0.01) # Stochastic gradient descent, learning rate of 0.01

for epoch in range(500):
    y_pred = model(x_data)
    
    loss = criterion(y_pred, y_data) # create loss for your data
    print(epoch, loss.item()) # print your epoch and loss
    optimizer.zero_grad() # zero your gradient function
    loss.backward() # back propogation
    optimizer.step() ## readjust your optimizer
    
output = Variable(torch.Tensor([[4.0]])) # define a new variable to test
y_pred = model(output)

print("my prediction after the training", 4, model(output).data[0][0])
```
